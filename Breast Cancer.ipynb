```python
def sigmoid(X):
    return 1.0/(1.0 + np.exp(-X))
     
```

```python
def hypothesis(X, theta):
    return sigmoid(np.dot(X, theta))
```

```python

def cost_function(X, Y, theta):
    h_theta = hypothesis(X, theta)
    cost = np.mean(Y*np.log(h_theta) + (1 - Y)*np.log(1 - h_theta))
    cost = cost*(-1)

    return cost
```

```python

def gradient(X, Y, theta):

  h_theta = hypothesis(X, theta)

  grad = np.dot(X.T, (Y - h_theta))

  return grad/X.shape[0]
```

```python
def gradient_descent(X, Y, learning_rate = 0.001, max_steps = 1000):
     m, n = X.shape
     theta = np.zeros((n, 1))
     cost_epoch = []

     for i in range(max_steps):

        grad = gradient(X, Y, theta)

        e = cost_function(X, Y, theta)
        cost_epoch.append(e)

        theta = theta + learning_rate*grad

     return (theta, cost_epoch)
```

```python
theta,cost_epoch=gradient_descent(X_train,y_train,0.1,200)
```

```python

def predict(X, theta):
    h_theta = hypothesis(X, theta)
    output = np.zeros(h_theta.shape)

    output[h_theta>0.5] = 1
    output = output.astype('int')

    return output
```

```python
train_preds = predict(X_train, theta)
test_preds = predict(X_test, theta)
test_preds.shape
```

```python
def accuracy(preds,labels):
    labels=labels.astype('int')
    print(np.sum(preds==labels)/labels.shape[0])
```
