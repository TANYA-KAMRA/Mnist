```python
Y_train=(Y_train==1).astype(int)
Y_test=(Y_test==1).astype(int)
```

```python
ones=np.ones((X_train.shape[0],1))
X_train=np.hstack((ones,X_train))
```

```python
ones=np.ones((X_test.shape[0],1))
X_test=np.hstack((ones,X_test))
```

```python
Y_train=np.array(Y_train)
```

```python
Y_train=Y_train.reshape(-1,1)
```

```python
Y_test=np.array(Y_test)
Y_test=Y_test.reshape(-1,1)
```

```python
def sigmoid(X):
    return 1.0/(1.0 + np.exp(-X))
```

```python
def hypothesis(X, theta):
    return sigmoid(np.dot(X, theta))
     
```

```python
def cost_function(X, Y, theta):
    h_theta = hypothesis(X, theta)
    cost = np.mean(Y*np.log(h_theta) + (1 - Y)*np.log(1 - h_theta))
    cost = cost*(-1)

    return cost
```

```python

def gradient(X, Y, theta):

  h_theta = hypothesis(X, theta)

  grad = np.dot(X.T, (Y - h_theta))

  return grad/X.shape[0]
```

```python
def gradient_descent(X, Y, learning_rate = 0.001, max_steps = 500):

  m, n = X.shape
  theta = np.zeros((n, 1))
  cost_epoch = []

  for i in range(max_steps):

    grad = gradient(X, Y, theta)

    e = cost_function(X, Y, theta)
    cost_epoch.append(e)

    theta = theta + learning_rate*grad

  return (theta, cost_epoch)
```

```python
theta,cost_epoch=gradient_descent(X_train,Y_train,0.1,1000)
```

```python
def predict(x,theta):
    y_p=hypothesis(x,theta)
    
    output=np.zeros(y_p.shape)
    
    output[y_p>0.5]=1
    output=output.astype('int')
    return output
```

```python
test_preds=predict(X_test,theta)
train_preds=predict(X_train,theta)
print(test_preds)
```
